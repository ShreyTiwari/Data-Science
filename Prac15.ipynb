```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats
from sklearn.metrics import r2_score

ds = pd.read_csv('Residuals.csv')

x = ds['Wind Speed']
y = ds['Current']

b, a = np.polyfit(x, y, 1)
yp = a + b*np.array(x)

plt.plot(x, y, 'r.')
plt.plot(x, yp)
plt.show()


print("\n", scipy.stats.pearsonr(x, y))
print("\nr2 : ", r2_score(y, yp))       #takes parameters as observed, predicted.
plt.scatter(y, yp, color = "red") 
plt.plot(y,y)
plt.xlabel("Observed y")
plt.ylabel("Predicted y")
plt.show()
#Residual Plots resi = y - yp plt.plot(yp, resi, "rx") plt.plot([1,4], [0,0]) plt.show()
```
