```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t
from scipy.stats import norm

x = np.arange(-6,6.2,1/20)

plt.plot(x, (norm.pdf(x)), color="blue")  #blue curve
plt.plot(x, (t.pdf(x , 3)), color="red") #red curve
plt.plot(x, (t.pdf(x , 20)), color="green")#green curve

plt.xlabel("Z-Score")
plt.ylabel("Prob Density Function Values")
plt.title("Comparison between t-distribution and normal curve")
plt.show()
```
