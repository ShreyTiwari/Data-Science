```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.stats import geom

def Geo(n, p):
    pmf=[]
    for x in range(1,n+1):
        pmf.append(geom.pmf(x, p))
    plt.bar(range(1,n+1), pmf, width =1, align = 'center')
    plt.xticks(range(1,n+1))
    plt.show()
    print(pmf)
    
Geo(10,0.6)
Geo(100,0.01)
Geo(10,0.9)

#Geom takes two parameters geom(trial no on which first success is expected, probability of success)
```
