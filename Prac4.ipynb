```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ds = pd.read_csv("Cleaned.csv")

plt.scatter(ds.Height, ds.Weight)
plt.xlabel("Height")
plt.ylabel("weight")
plt.title("Slight relationship")
plt.show()
```

```python
sns.swarmplot(ds.PL, ds.CGPA)
plt.show()

#swarmplot requires one axis to be qualitative and the other to be quantitative.
#Whereas plt.scatter needs both the axis to be quantitative
```

```python
def Convert(x):
    if(x == 'Yes'):
        x = 1
    else:
        x = 0
    return x

ds = pd.read_csv('Cleaned.csv')

array = ds.Sim_Course.apply(Convert)

sns.swarmplot(array, ds.Focus)
plt.show()
```
