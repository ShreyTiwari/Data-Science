```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm


#norm.ppf gives the z-score corresponding to the area entered (entered area is area covered by z - score to its left)
#norm takes either 1 or 3 parameters....1->area and assumes that mean = 0 and std = 1.
#                                       3->area, mean, std of the distribution.


def normCheck(data):
    Q = []
    data = np.sort(np.array(data))
    Q = [norm.ppf((i-0.5)/len(data), np.mean(data), np.std(data, ddof = 1)) for i in range(1, len(data)+1)]
    plt.plot(data, Q, 'ro');
    plt.plot(data, data, "green");
    plt.show()
    

x=[3.01, 3.35, 4.79, 5.96, 7.89]
normCheck( x ) 
```
