```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

ds = pd.read_csv("Cleaned.csv")
df = pd.read_csv("Class_Data.csv")

ds.Height.hist(color='red')
plt.title("Height Histogram")
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.show()

print("Mean = ",np.mean(ds.Height))
print("Median = ", np.median(ds.Height))
print("Almost Symmetric. [Mean equivalent to median]")

print("\nOther Examples")
df.Height.hist(color='blue')
plt.title("Height Histogram")
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.show()


ds.CGPA.hist(color="black", histtype="step")
plt.show()
print("Negatively Skewed")

print("Mean = ", np.mean(ds.CGPA))
print("Median = ", np.median(ds.CGPA),"\n")


```

```python
plt.hist(ds.CGPA, bins = [i for i in range(11)])
plt.show()

plt.hist(ds.CGPA, bins = 10, color="red", histtype="step")
plt.show()

print("Slight difference .... Better to use system generated bins if specific bins arent required\n")

print("plt.hist() returns frequency, bins and patches")
f, bins, patches = plt.hist(ds.CGPA, bins = 10)
print("\nfrequency = ", f)
print("bins = ", bins)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics as st
import seaborn as sns
from pandas.tools.plotting import scatter_matrix

df = pd.read_csv('Class_Data.csv')

print("To Calculate BMI of the Data Set:\n")

h_median = np.median(df.Height) 
for index, row in df['Height'].iteritems():
    if row < 10:
        df['Height'].loc[index] = row*30.48
    if row < 125:
        df['Height'].loc[index] = h_median

W_median = np.median(df.Weight)
for index, row in df['Weight'].iteritems():
    if(row < 35):
        df['Weight'].loc[index] = W_median
        
bmi = df.Weight/pow(df.Height/100,2)
print(np.mean(bmi))


df = df.assign(BMI = bmi)
print(df.columns)

#del df.bmi

df.to_csv('Cleaned.csv')
```
