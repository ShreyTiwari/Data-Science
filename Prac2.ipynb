```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as st
import matplotlib.pyplot as plt

ds = pd.read_csv("Cleaned.csv")

ds.Gender.value_counts().plot(kind="bar")
plt.show()

x = ds.bday.value_counts()
plt.bar(range(len(x)), x.get_values())
plt.show()

ds.bday.value_counts().plot(kind = "bar")
plt.show()

ds.Focus.value_counts().plot(kind = "bar")
plt.show()
```

```python
sns.barplot(data=ds, x='PL', y='CGPA', hue='Gender', ci = 0)
plt.show()
```

```python
colors = ['red', 'blue']

ds.Gender.value_counts().plot(kind = 'pie',explode=(0,0.1), autopct = '%1.2f%%', colors=colors)
plt.axis('equal')
plt.show()

# with the percent listed as a fraction
#autopct='%1.1f%%',
```

```python
# Create a list of colors (from iWantHue)
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]

# Create a pie chart
plt.pie(
    # using data total)arrests
    ds['CGPA'],
    # with the labels being officer names
    labels=ds['PL'],
    # with no shadows
    shadow=False,
    # with colors
    colors=colors,
    # with one slide exploded out
    explode=(0, 0, 0, 0, 0.15),
    # with the start angle at 90%
    startangle=90,
    # with the percent listed as a fraction
    #autopct='%1.1f%%',
    )

# View the plot drop above
plt.axis('equal')

# View the plot
plt.tight_layout()
plt.show()



#link : https://chrisalbon.com/python/matplotlib_pie_chart.html
```
