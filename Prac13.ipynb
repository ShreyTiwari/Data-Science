```python
%matplotlib inline
import matplotlib.pyplot as plt
from math import sqrt
from scipy.stats import norm

# H0: Mu >= 50 vs H1: Mu < 50.

Mu0 = 50
n = 30
alpha = 0.05
sigma = 10
```

```python
#1. Moving away from the Null Distribution:

z_nd = norm.ppf(alpha)
se = sigma / sqrt(n)

cxb = z_nd * se + Mu0            #critical x bar value

power = []

for alt in range(40, Mu0):
    a = (cxb-alt)/se
    power.append(norm.cdf(a))
    
plt.plot(range(40, Mu0), power, color = "orange")
plt.show()
```

```python
# Sample size increases Power increases

n = [10, 30, 50, 100, 500]

for x in n:
    z_nd = norm.ppf(alpha)
    se = sigma/sqrt(x)
    cxb = z_nd * se + Mu0
    power = []
    for alt in range(40, Mu0):
        a = (cxb - alt)/se
        power.append(norm.cdf(a))
    label_val = "n = "+str(x)
    plt.plot(range(40, Mu0), power, label = label_val)
    plt.legend(loc = 'lower left')
plt.show()
```

```python
# Level of test increases Power increases

alpha_list = [0.01, 0.05, 0.10]
for alpha in alpha_list:
    z_nd = norm.ppf(alpha)
    se = sigma /sqrt(30)
    cxb = z_nd * se + Mu0
    power = []
    for alt in range(40, Mu0):
        a = (cxb - alt)/se
        power.append(norm.cdf(a))
    label_val = "alpha = "+str(alpha)
    plt.plot(range(40, Mu0), power, label = label_val)
    plt.legend(loc = 'lower left')
plt.show()
```

```python
# Sigma increases Power decreases

sigma_list = [5, 10, 15, 20]

for sigma in sigma_list:
    z_nd = norm.ppf(alpha)
    se = sigma /sqrt(30)
    cxb = z_nd * se + Mu0
    power = []
    for alt in range(40, Mu0):
        a = (cxb - alt)/se
        power.append(norm.cdf(a))
    label_val = "sigma = "+str(sigma)
    plt.plot(range(40, Mu0), power, label = label_val)
    plt.legend(loc = 'lower left')
plt.show()
```
