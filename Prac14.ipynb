```python
%matplotlib inline
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt

x = np.array([284.5, 407.5, 512, 551, 578.5, 697, 746.5])
y = np.array([1255, 3390, 5560, 6670, 7160, 9820, 11020])

b, a = np.polyfit(x, y, 1, full=False)

z = [(a+b*i) for i in x]

plt.plot(x,z)
plt.plot(x, y, 'r.')
plt.show()

residuals = y - z

#print(residuals)
#print('\n',sum(residuals))

plt.plot(z, residuals, 'bo')
plt.title("We Should Not Find Any Correlation")
plt.show()

print("\n",np.corrcoef(x,y), "\n") # R_{ij} = frac{ C_{ij} }/{sqrt{ C_{ii} * C_{jj} } }
print(scipy.stats.pearsonr(x,y))   # Calculates a Pearson correlation coefficient and the p-value for testing non-correlation.
```
