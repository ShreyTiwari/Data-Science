```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm
from math import sqrt

def pos(rate, a, b, x1, x2):
    array1 = np.arange(a, b, 1)
    array2 = np.arange(a, b, 1/20)
    
    plt.figure(figsize=(10,5))
    plt.title("Poisson")
    plt.bar(array1, poisson.pmf(array1, rate), align="center", fill=False)
    plt.plot(array2, norm.pdf(array2, rate, sqrt(rate)))             #remember that norm function always takes S.D not var.
    section = np.arange(x1, x2, 1)
    plt.fill_between(section, poisson.pmf(section, rate))
    plt.show()
    
pos(81, 45, 120, 89, 120)
pos(27, 5, 50, 30, 50)        #one point given as x1 = x2.
```

```python
#P(X >=90):
#Not in tail region so CC helps get better appoximation.


pos(81, 40, 150, 89, 150)


print('Using Poisson , X ~ Poisson(27 * 3) : ', 1 - poisson.cdf(89, 81))

print('Using Normal, X ~ N(81, sd = 9) : ', norm.sf(90, 81, 9))

print('Continuity Correction applied', norm.sf(89.5, 81, 9))
```

```python
#P(X >=42):
#In the tail region so effect of CC not significant as it fails to account skewness in the tails.


pos(27, 0, 70, 42, 100)



print('Using Poisson , X ~ Poisson(27) : ', 1 - poisson.cdf(41, 27))

print('Using Normal, X ~ N(27, sd sqrt(27)) : ', norm.sf(42, 27, sqrt(27)))

print('Continuity Correction applied', norm.sf(41.5, 27, sqrt(27)))
```
